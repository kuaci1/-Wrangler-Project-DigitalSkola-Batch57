## Map & replace

```python
df.columns
```

```python
#mengubah isi kolom menjadi semua huruf kecil
df['job_title'] = df['job_title'].str.lower()
df['job_title']
```

Refrence / master *data*

```python
dict_experience_level = {
'SE' : 'Senior',
'MI': 'Middle',
'EN' : 'Entry',
'EX' :'Executive'}

df['experience_level'] = df['experience_level'].replace(dict_experience_level)
df['experience_level']
```

```python
df['employment_type'] = df['employment_type'].replace(
{
'FT' : 'Full Time',
'CT' : 'Contract',
'FL' : 'Freelance',
'PT' : 'PartTime'
}
)
df['employment_type']
```

```python
df['remote_ratio'].unique()
```

```python
df['remote_category'] = df['remote_ratio'].copy()
df['remote_category'] = df['remote_category'].replace(
  {0:'offline',
  50:'hybrid',
  100:'remote'}
)
```

```python
# 3. Cari Gaji Tertinggi di antara pekerja remote
# idxmax() akan memberikan index baris dengan nilai maksimum
pekerja_remote = df[df['remote_category'] == 'remote']
index_gaji_max = pekerja_remote['salary_in_usd'].idxmax()
data_tertinggi = pekerja_remote.loc[index_gaji_max]

print(f"Gaji maksimum untuk pekerja remote adalah: {data_tertinggi['salary_in_usd']}")
print(f"Employee Residence: {data_tertinggi['employee_residence']}")
print(f"Job Title: {data_tertinggi['job_title']}")
```

```python
#5. Tunjukkan ada berapa banyak pekerja per company location
df['company_location'].value_counts()
```

```python
#4. Analisa pekejra dari employee_residence apakah sudah lebih dari UMR indonesia (IN) (5,7 juta)
# kalau ada yang dibawahnya tampilkan datanya
df_in = df[df['employee_residence']=='IN'].copy()
#menentukan permisalan angka kurs
umr_indo = 5000000
kurs_usd = 15500

df_in['salary_in_idr'] = (df_in['salary_in_usd'] * kurs_usd) / 12
diatas_UMR = df_in[df_in['salary_in_idr'] > umr_indo]
dibawah_UMR = df_in[df_in['salary_in_idr'] <= umr_indo]

print("HASIL ANALISA GAJI (IN)")
print(f"Jumlah pekerja DI ATAS UMR : {len(diatas_UMR)} orang")
print(f"Jumlah pekerja DI BAWAH UMR: {len(dibawah_UMR)} orang")

gaji_terendah = df_in['salary_in_idr'].min()
print(f"Gaji Paling Kecil: Rp {gaji_terendah:,.0f}")
```
